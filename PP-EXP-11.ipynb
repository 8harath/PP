```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({'Year': [2020, 2021, 2022], 'Sales': [100, 150, 200]})

plt.plot(df['Year'], df['Sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales Over Years')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Fruits': ['Apple', 'Banana', 'Cherry'], 'Count': [10, 15, 7]}
df = pd.DataFrame(data)

plt.bar(df['Fruits'], df['Count'])
plt.xlabel('Fruits')
plt.ylabel('Count')
plt.title('Fruit Count')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({'Date': pd.date_range('20220101', periods=5),
                   'Value': [10, 15, 20, 25, 30]})

plt.plot(df['Date'], df['Value'])
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Data')
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
values = [5, 7, 3]

plt.bar(categories, values)
plt.title('Bar Chart Example')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({'Category': ['X', 'Y'],
                   'Value1': [10, 15],
                   'Value2': [20, 25]})

plt.bar(df['Category'], df['Value1'], label='Value1')
plt.bar(df['Category'], df['Value2'], bottom=df['Value1'], label='Value2')

plt.legend()
plt.title('Stacked Bar Chart')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})

corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
